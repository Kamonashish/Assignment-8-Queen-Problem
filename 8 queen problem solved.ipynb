```python
no_of_queens = 8  

c = [0 for x in range(no_of_queens)]  
a = []

def isSafe(test_row,test_column):
    if test_row == 0:
        return True

    for row in range(0,test_row):
        if test_column == c[row]:
            return False
        if abs( test_row - row ) == abs(test_column - c[row]):
            return False
    return True

for queen0 in range(no_of_queens):
    if not isSafe(0,queen0):
        continue
    else:
        c[0] = queen0

    for queen1 in range(no_of_queens):
        if not isSafe(1,queen1):
            continue
        else:
            c[1] = queen1

        for queen2 in range(no_of_queens):
            if not isSafe(2,queen2):
                continue
            else:
                c[2] = queen2

            for queen3 in range(no_of_queens):
                if not isSafe(3,queen3):
                    continue
                else:
                    c[3] = queen3

                for queen4 in range(no_of_queens):
                    if not isSafe(4,queen4):
                        continue
                    else:
                        c[4] = queen4

                    for queen5 in range(no_of_queens):
                        if not isSafe(5,queen5):
                            continue
                        else:
                            c[5] = queen5

                        for queen6 in range(no_of_queens):
                            if not isSafe(6,queen6):
                                continue
                            else:
                                c[6] = queen6

                            for queen7 in range(no_of_queens):
                                if not isSafe(7,queen7):
                                    continue
                                else:
                                    c[7] = queen7
                                    a.append( c.copy() )

print(len(a), "possible solutions found")
for solution in a:
    print(solution)
```
